```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('imdb-videogames.csv')

data = data.replace(True, 1)
data = data.replace(False, 0)
data = data.replace("E", 1)
data = data.replace("GA", 1)
data = data.replace("K-A", 1)
data = data.replace("EC", 1)
data = data.replace("G", 1)
data = data.replace("Unrated", 1)
data = data.replace("Not Rated", 1)
data = data.replace("Approved", 1)
data = data.replace("Passed", 1)
data = data.replace("CE", 1)
data = data.replace("E10+", 2)
data = data.replace("PG", 2)
data = data.replace("T", 2)
data = data.replace("TV-14", 2)
data = data.replace("MA-13", 2)
data = data.replace("PG-13", 2)
data = data.replace("M", 3)
data = data.replace("MA-17", 3)
data = data.replace("AO", 3)
data = data.replace("TV-MA", 3)


data = data.fillna(0)

# Extract relevant columns
features = ['certificate','Action','Adventure','Comedy','Crime','Family','Fantasy','Mystery','Sci-Fi','Thriller']
scaled = data[['certificate', 'Action','Adventure','Comedy','Crime','Family','Fantasy','Mystery','Sci-Fi','Thriller']].values
data
```

```python
scaled_data = data.copy()
scaler = StandardScaler().set_output(transform = "pandas")
scaled_features = scaler.fit_transform(data[features])
scaled_data[features] = scaled_features
scaled_data
```

```python
pca = PCA(n_components=2)
pca_df = pca.fit_transform(scaled_features)
pca_df
```

```python
def get_euclidean_dist(target_game, k=10):
    similarities = []
    for i, row in scaled_data.iterrows():
            euclidean_similarities = euclidean_distances([target_game], [scaled[i]])
            similarities.append((row['name'], euclidean_similarities[0][0]))
    return sorted(similarities, key=lambda x: x[1], reverse=True)[:k]
```

```python
game = scaled_data[scaled_data['name'] == 'Red Dead Redemption II'][features].values[0].tolist()
get_euclidean_dist(game)
```

```python
game_2 = scaled_data[scaled_data['name'] == "Xenoblade Chronicles 3"][features].values[0].tolist()
get_euclidean_dist(game_2)
```
